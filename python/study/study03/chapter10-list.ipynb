```python
from re import search
# 슬라이싱
로또 = [3 , 11 ,  23 , 33 , 43 , 45]

#로또 변수에 있는 리스트의 인덱스 1번부터 2번까지의 값을 출력해
print(로또[1:3])

# 로또 변수의 마지막 위치 출력해줘
print(로또[-1])

# 반대로 출력
print(로또[::-1])

# 로또 리스트 안에 33이 있는지 확인해보세요
print('33' in 로또) # 값 in 변수 = True or False 로 확인가능

로또2 = [2,12,15,24,33,39]
#로또 리스트를 로또 리스트2와 합치기
print(로또 + 로또2)


# range를 이용해서 1부터 10사이에 짝수만 들어있는 짝수(변수명) 리스트를 만들어주세요
짝수 = list(range(2,12,2))
print(짝수)
# 짝수 리스트에 들어있는 값을 2배 늘리고 다시 짝수 수에 담아줘
# ran = len(짝수)
# for i in range(ran) :
#     짝수[i] = 짝수[i]*2

# for 반복문을 통해서 담겨진 값을 실제 2배하는 방법도 가능하다.

짝수 = 짝수 * 2
print(짝수)
#짝수 리스트에 있는 요소의 개수를 구해줘
print(len(짝수))
#짝수 리스트에 3번째 인덱스를 출력해줘
print(짝수[3])
#len() 함수를 이용해 인덱스 마지막 값을 출력해줘
print(짝수[len(짝수)-1])
```

```python
# list 값 넣고 , 바꾸고 , 지우고

mzfood = ["숙주" , "분모자" , "마라" , "소세지" , "소고기", "옥수수면"]

#값을 넣는 방법 2가지 : append(): 마지막에 요소에 값을 넣음 , insert() : 원하는 곳에 값을 넣음

mzfood.append("고수")
print(mzfood)

mzfood.insert(0, "탕후루")
print(mzfood)

# 값 지우기
del mzfood[-1]
print(mzfood)

del mzfood[0]
print(mzfood)


#값 바꾸기
mzfood[-2] = "양고기"
print(mzfood)

# 문자열 시퀀스 구조 아닌가?
인사 = "반가워"

# 인사[2] = "웡"  'str' object does not support item assignment 리스트형이 아니라 할당 불가능

print(dir(list))

```

```python
print(mzfood)

mzfood.reverse() # 반대로 담기
print(mzfood)

mzfood.sort() # 순서대로 담기
print(mzfood)

# reverse 왜 print 문을 쓰면 none 나올까?
print(mzfood)
print(mzfood.reverse())
print(mzfood.sort()) # 실제로 코드 실력은 됐으나 변수에 담지 않아 사라지고 출력결과 값이 없기때문에 None 이 출력됨.
```
