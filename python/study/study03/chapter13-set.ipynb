* set
        * set은 수학의 집합을 의미한다.
        * 딕셔너리와 같이 순서가 없다.
        * 값의 중복을 허락하지 않는다. 매우중요!
        * set 생성 방법도 중괄호
        

```python
채소 = {"당근", "배추", "대파" , "양파"}
type(채소)
print(채소)

# 순서가 없고, 출력하면 요소가 다르게 나온다는 특징이 있다.

# 세트를 만드는 다양한 방법

세트= set()
세트

세트2 = {}
세트2

```

```python
# 중복을 제일 싫어하는 set

사과_리스트 = list("apple")
print(사과_리스트 )
사과_튜플 = tuple("apple")
print(사과_튜플)
사과_set = set("apple")
print(사과_set)
```
