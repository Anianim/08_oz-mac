* dict(딕셔너리)
        * list, tuple , str , range (범위안에 연속된 데이터를 만드는 시퀀스)
        공통적인 특징을 가지고 있음 데이터의 연관성이 없다.
        * dict는 여러개의 값이 일렬로 정렬되면서도 값끼리의 연관성이 존재
        (ex)게임캐릭터 힘:30 , 지능:20 , 체력:40 , 민첩:70
        * dict 생성방법은 중괄호


```python
# list [] , tuple () , dict {}

전사 = {"힘":30 , "지능":20 , "체력":40 , "민첩":70}
전사

# { 키:값 } # key : value 형태로 넣어줘야함
# { key : value, key2 : value2 }
# 시퀀스 데이터는 인덱싱을 이용해 데이터를 찾았습니다.

# 전사[1]
전사["힘"]
print(f'전사님은 힘이 {전사["힘"]}입니다.')
```

```python
# 딕셔너리 만드는 다양한 방법

#빈 딕셔너리 만들기

딕셔너리 = {}
딕셔너리
print(type(딕셔너리))


딕셔너리2 = dict()
딕셔너리2
print(type(딕셔너리2))

#값을 넣어 만드는 방법
#dict()로 생성시 값 넣는 기호 key=value  /  "key":value 아님!  주의 필요
딕셔너리3 = dict(힘=30, 지능=20, 체력=40, 민첩=90) #형태 꼭 기억하기
print(딕셔너리3)

# 딕셔너리 만드는 다양한 방법

#빈 딕셔너리 만들기

딕셔너리 = {}
딕셔너리
print(type(딕셔너리))


딕셔너리2 = dict()
딕셔너리2
print(type(딕셔너리2))

#값을 넣어 만드는 방법
#dict()로 생성시 값 넣는 기호 key=value  /  "key":value 아님!  주의 필요
딕셔너리3 = dict(힘=30, 지능=20, 체력=40, 민첩=90) #형태 꼭 기억하기
print(딕셔너리3)

#zip() : 병렬처리를 도와준다.

딕셔너리4 = dict(zip(["힘", "지능", "체력", "민첩"], [30,20,40,80])) #배열의 개수를 맞춰줘야함 개수 안맞을시 작은 배열기준으로 잘라버림
딕셔너리4
```

```python
#dict 이력서를 작성했는데 붕어빵 인터네셔널 채용공고 열정만 있으면 오세요
# 바로오 갔더니 코드만 있었다 .. 레거시
#리팩토리 해라 함.

주문_개수 = map(int , input(f'붕어빵 맛 순서대로 개수 입력해주계소 안살꺼면 0입력 (팥, 슈크립, 잡채)').split(','))
제품_가격 = {"팥": 2000, "슈" : 2500 , "잡" : 3000}

주문_개수 = list(주문_개수)
주문_개수_합 = sum(list(주문_개수))
total = 0

total += 제품_가격["팥"] * 주문_개수[0]
total += 제품_가격["슈"] * 주문_개수[1]
total += 제품_가격["잡"] * 주문_개수[2]


print(f'주문이 완료되었습니다. 팥{주문_개수[0]}개, 슈크림{주문_개수[1]}개, 잡채{주문_개수[2]}개 총 {주문_개수_합}개 결제 금액은 {total}원 입니다.')
```
