### 시퀀스 자료형
* 문자열 :
        * 시퀀스형에서 사용할수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱 ( [:] [: :] )
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)

* list :
        * 시퀀스형에서 사용할수 있는 메소드(count(), index(), split() 등...)
        * 슬라이싱 ( [:] [: :] )
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
        * list 지원하는 다양한 메소드(append(), insert(), sort() , reverse() 등등)
    
* tuple :
        * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제할 수 없다.
        * 값을 추가, 삭제, 변경할 수 없기 때문에, 사용 못하는 메소드가 많다.
        * tuple의 생성 방법은 tuple = 값, 값 또는 tuple = (값 , 값)
        * 튜플은 소괄호 입니다.
        

```python
# 리스트 대괄호 [] , 튜플은 소괄호 ()
튜플 = (1, "이" , 3.0)
print(튜플)
print(type(튜플))

# 리스트 = [1, "이" , 3.0]
# print(type(리스트))
# print(리스트)

튜플2 = 1, "이" , 3
print(type(튜플2))
print(튜플2)

#하나의 값만 들어있는 튜플 만들기
튜플_솔로 = (1, )
print(튜플_솔로)
print(type(튜플_솔로))

튜플_솔로2 = (1)
print(튜플_솔로2)
print(type(튜플_솔로2)) #1개의 값이면 튜플로 인정 안함. 정수 들어가서 int 타입이 출력됨.
```

```python
# 값 추가, 삭제, 변경 불가.... 진짜 그럴까?
print(튜플)
print(dir(튜플))

# print(dir(list))


튜플_리스트화 = list(튜플)
print(type(튜플_리스트화)) #list 타입으로 형변환됨

```

```python
# 번외

수강생 = "중꺽마"

수강생_list = list(수강생)
수강생_tuple = tuple(수강생)

print(수강생_list)
print(수강생_tuple)

수강생_list[2] = "태"
print(수강생_list)

# join함수 (알고 있으면 코테 도움됨)
취뽀_수강생 = "".join(수강생_list) #배열 타입의 데이터를 합쳐주는데 "".join 은 아무 데이터 삽입 없이 합치고 "-".join 넣을경우 데이터 사이에 "-"를 삽입해서 합쳐준다.
print(취뽀_수강생)
취뽀_수강생 = "-".join(수강생_list)
print(취뽀_수강생)

```

```python
# 공통점은 무엇일까?
print(튜플)
#길이를 구할 수 있다.
print(len(튜플))
#곱하기 연산으로 요소를 복사 할 수있따.
print(튜플*3)
#튜플은 요소들 끼리 더할 수 있다. (내용적인 측면 계산X)
print(튜플+튜플)
#튜플은 인덱싱이 가능하다
print(튜플[2])
#튜플의 인덱스
print(튜플.index(3.0))
#튜플 요소의 개수
print(튜플.count(1))
#튜플도 슬라이싱이 가능하다
튜플[1:3]
```
