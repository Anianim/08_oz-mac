##사칙연산
더하기 : +

빼기 : -

곱하기 : *

나누기 :
*  /  : 우리가 알고 있는 나누기랑 동일한 기능 (예 5/2 = 2.5)
*  // : 몫을 반환 (예 5/2 = 2)
*  %  : 나머지를 반환 (예 5/2 = 1)

거듭제곱 : **


```python
# 더하기
print(3+2)
```

```python
# 빼기
print(3-1)
```

```python
# 곱하기
print(3*2)
```

```python
# /
print ( 7 / 2 )

# //
print( 7 // 2 )

# %
print ( 7 % 2)
```
