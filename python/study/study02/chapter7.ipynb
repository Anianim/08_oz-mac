```python
숫자 = "123456789"
#전체 출력
print(숫자[::])

#전체 반대로 출력
print(숫자[::-1])

#첫번째 인덱스 기준으로 2칸씩 띄어쓰기 출력하기
print(숫자[::2])

#3번째 인덱스 기준으로 2칸씩 띄어쓰기 출력하기
print(숫자[4:1:-1]) # 4번째 인덱스에서 1번쨰 인덱스까지 출력 역순이기때문에 -1 이 필요함
print(숫자[1:4]) # 1번째 인덱스 에서 4번째 인덱스 까지 출력 정순이기 떄문에 -1 필요없음
```
