```python
won = {'A등급':1000, 'B등급':500, 'C등급':100}

def 계산(a,b,c):
    합계 = won['A등급']* a + won['B등급']*b + won['C등급']*c
    return 합계
# dict 타입 한해서만 전역변수로 부터 값을 가져와 사용 사능하다.
print(계산(5,2,3), '원입니다')
```

```python

A = 0
B = 0
C = 0
유물 = 0

def 감자():
    global A
    A += 1

def 양파():
    global A, B, C
    A += 3
    B += 3
    C += 4

def 옥수수():
    global 유물
    유물 += 1

감자()
양파()
옥수수()

합계 = A + B + C + 유물

print('오늘 수확한 수확물:', 합계)
```

```python
def 수확철():
    야채 = {'A등급': 0, 'B등급': 0, 'C등급': 0}

    def 양파():
        야채['A등급'] += 1

    def 대파():
        야채['A등급'] += 3
        야채['B등급'] += 3
        야채['C등급'] += 4

    def 쪽파():
        야채['금'] = 1

    양파()
    대파()
    쪽파()

    return 야채

print(수확철())
#print(sum(수확철().values()))
```
