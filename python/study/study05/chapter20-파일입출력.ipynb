```python
f = open('test.txt','w')
메세지 = ''

for i in range(10):
    메세지 += "개발자 여러분\n"
f.write(메세지)
f.close()
```

```python
f = open('test.txt','r')
s = f.read()
print(s)
f.close()
```

```python
f = open('test.txt','r')
line = f.readline()
print(line)
print(type(line))
f.close()
```

```python
f = open('test.txt','r')
while line :
    line = f.readline()
    print(line)
f.close()
```
