```python
#1. while문 만들어보기

i = 0 #초기식
while i <10 : #while 조건식
    print('점점 성장하는 여러분') #반복할 코드
    i += 1 #변환식
```

```python
# 3.while문 조건식 증감, 감소

#1. 기본적인 증가 유형

i=1
while i < 10 :
    print('이정도로 성장할줄이야',i)
    i += 1

i=10
while i > 0 :
    print('이정도로 성장할줄이야2',i)
    i -= 1
```

```python
#5. while문이 for보다 강점인 부분?
# 무수히 많이 시행해야하는 경우 강점이 있음
import random
random.randint(1,45)

i= 0
while i != 4:
    i = random.randint(1,6)
    print(i)

```

```python
#6 while 문에도 else 가 있다.

양파  = 1

while 양파 < 10:
    print("양파", 양파, "개 남았습니다.")
    if 양파 == 0:
        break
    양파 += 1
else:
    print("양파가 다 떨어졌습니다.") #while문이 끝날때 실행됨
```

```python
#7. 반복문에 흐름을 제어하는 continue와 break

i=0
while True:
    print(i)
    i+=1
    if i == 10 :
        break

# for i in range(10):
#     print(i)
#     if i == 10 :
#         break

```

```python
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    print(i)

for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
#8. 붕어방 인터네셔널은 내가살린다.
# for 문을 이용해 리팩토링을 해주세요 . input()
# 잡담 게시판에 여러분의 코드를 자랑해주세요

붕어빵= {
        "붕어빵_팥" : 0 ,
        "붕어빵_슈" : 0 ,
        "붕어빵_잡" : 0
    }
붕어빵가격 = [ 2000 , 2500 , 3000 ]
서비스 = 0
총액 = 0

while True :

    서비스 = int(input('원하는 서비스를 선택해주세요 1.팥붕 , 2.슈붕 , 3.잡붕 4.결제'))
    if 서비스 == 1 :
        i= int(input('팥붕 원하시는 개수를 입력해주세요'))
        print(f'팥붕 {i}개 선택 했습니다.')
        붕어빵["붕어빵_팥"] += i

    elif 서비스 == 2:
        i= int(input('슈붕 원하시는 개수를 입력해주세요'))
        print(f'슈붕 {i}개 선택 했습니다.')
        붕어빵["붕어빵_슈"] += i
        continue
    elif 서비스 == 3:
        i= int(input('잡붕 원하시는 개수를 입력해주세요'))
        print(f'잡붕 {i}개 선택 했습니다.')
        붕어빵["붕어빵_잡"] += i
        continue
    else :
        print(f'팥붕어빵 {붕어빵["붕어빵_팥"]}개, 슈붕어빵 {붕어빵["붕어빵_슈"]}개, 잡붕어빵 {붕어빵["붕어빵_잡"]}개')
        print(f'결제금액은 {붕어빵["붕어빵_팥"]*붕어빵가격[0]+붕어빵["붕어빵_슈"]*붕어빵가격[1]+붕어빵["붕어빵_잡"]*붕어빵가격[2]}원 입니다.')
        print('감사합니다 또 찾아주세요')
        break;
```

```python
#9. 예제로 while문 익혀보기

#콜라츠하는 사람은 주어진 수가 1이 될때까지 다음 작업을 반복하면, 모든 수를 1로 만들수 있다고 추측했다
#작업
#1-1 입력된 수가 짝수라면 2로 나눈다.
#1-2 입력된 수가 홀수라면 3을 곱하고 1을 더한다.
#3. 위 결과로 나온 수에 같은 작업을 1이 될 때까지 반복한다.
# 예를 들어, 입력된수가 6이라면 6-> 3->10->5->16->8->4->2->1

#num 변수에 있는 수를 1로 만드는데 몇번이나 작업을 반복한지 출력한다.

num = 7
정답 = 0

while num != 1:
    if num % 2 == 0:
        num = num/2
    else:
        num= num*3 +1
    정답 += 1

print(정답)
```
