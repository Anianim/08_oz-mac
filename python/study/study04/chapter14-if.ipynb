```python
#2. if문 문법과 주의사항(콜론, 들여쓰기)
개수 = int(input("붕어빵 몇개 구매하십니까?"))

if 개수 > 3 :
    print("개당 1800원 입니다.")
```

```python
#3. if문의 흐름(작동 원리)
# 코드는 위에 아래 순서대로 작동합니다.

# 2) 개수라는 변수에 담음 1) 입력을 받음
개수 = int(input("붕어빵 몇개 구매하십니까?"))
# 3) 개수 변수가 있네 값이 뭐야? 4 > 3 True  또는 False

if 개수 > 3 :
    print("개당 1800원 입니다.") # True 라면 if문의 내용이 실행됨.

if 개수 <= 3:
    print("개당 2000원 입니다")

```

```python
#5. if ~ else 문법과 주의사항(콜론, 들여쓰기)

개수 = int(input("붕어빵 몇개 구매하십니까?"))

if 개수 > 3 :
    print("개당 1800원 입니다.")
else:
     print("개당 2100원 입니다.")
```

```python
#6. if ~ else문 흐름 (작동원리)

개수 = int(input("붕어빵 몇개 구매하십니까?"))

if 개수 > 3 :
    print("개당 1800원 입니다.")
else:
     print("개당 2100원 입니다.")
```

```python
#7. 조건문 조건(True, False)
#True
if True:
    print(True)
else :
    print(False)

#False

if False:
    print(True)
else :
    print(False)

#None

if None:
    print(True)
else :
    print(False)

#0

if 0:
    print(True)
else :
    print(False)

#1

if 1:
    print(True)
else :
    print(False)

#1.1

if 1.1 :
    print(True)
else :
    print(False)

#'hello'

if 'hello':
    print(True)
else :
    print(False)

#''

if '':
    print(True)
else :
    print(False)

# not

if not True :
    print(True)
else :
    print(False)

# not ''

if not '' :
    print(True)
else :
    print(False)
```

```python
#8. 조건식 여러개 설정하기

a, b = 10, 30

if a == 10 and b == 30 :
    print("동일합니다.")
else:
    print("동일하지 않습니다.")

if a == 10 or b == 30 :
    print("동일합니다.")
else:
    print("동일하지 않습니다.")

if a >= 10 and a< 20 :
    print("10보다 크고 20보다 작습니다.")

if 10 <= a< 20 :
    print("10보다 크고 20보다 작습니다.")

```

```python
#9. elif ~ if 문

맛 = input("붕어빵 맛을 골라주세요(팥, 슈크림, 잡채)")

if 맛 == "팥":
    print(f'{맛}을 선택하셨습니다.')
elif 맛 == "슈크림":
    print(f'{맛}을 선택하셨습니다.')
else :
    print(f'{맛}을 선택하셨습니다.')
```
