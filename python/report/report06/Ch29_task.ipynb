주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
# 조기리턴이란? 함수는 return 을 만나면 종료하고 값을 리턴하는게 일반적이다. 이를 활용하여 특정 기능이 끝나면 return 으로 함수 종료하게 하는것.

def is_even(num):
    if num % 2 == 0:
        return True
    # if 구문은 True(참) 일경우에만 실행됨 해서 짝수가 아니라면 다음줄인 return False 가 작동하게 되고 짝수라면 조건문의 return True 가 실행 하게됨.
    # else를 사용하지 않아도 조기리턴을 이용할수 있다.
    return False

print(is_even(3))
```
