두 개의 정수를 받아 합을 반환하는 함수 add를 작성하세요. 매개변수와 반환값에 타입 어노테이션을 추가하세요.

######타입 어노테이션은 힌팅 프로세서이다. 실제 코드에 큰 영향을 주지는 않지만 이로인해서 프로젝트 팀원간에 도움을 줄수 있다.

```python
def return_number(num_a : int , num_b : int): # 어노테이션 > 변수명 : 데이터타입 으로 지정한다.
    sum = num_a + num_b
    return sum

print(return_number("1","2")) # 어노테이션을 지정해도 힌팅프로세서 일분 오작동으로 인정하지 않는다. 해서 문자열을 합친 "1"+"2" 로 출력됨.
print(return_number(1,2)) #정수를 받아 합계로 1+2 = 3 이 출력됨.
```
