팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
count = 0
def factorial(num):
    global count

    if num == 0:
        count += 1
        print("계산된횟수2= ", count)
        return 1

    else :
        count += 1
        print("계산된횟수= ", count)
        print("현재 num= ", num)
        return num * factorial(num - 1)

    # 5 * factorial(4)
    # 5 * 4 * factorial(3)
    # 5 * 4 * 3 * factorial(2)
    # 5 * 4 * 3 * 2 * factorial(1)
    # 5 * 4 * 3 * 2 * 1 * factorial(0)
    # 5 * 4 * 3 * 2 * 1 * 1
    #재귀함쉬는 for 이나 while 구문 대신 간결한 코드로 표현은 가능하지만, 메모리를 많이 차지하는 단점도 존재한다.

print(factorial(5))
```
