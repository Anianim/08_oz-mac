사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max(num_a ,num_b , num_c):
    numbers = num_a, num_b, num_c
    print(type(numbers))
    # 변수명 = 값 여러개 를 선언시 튜플이 만들어진다.
    print(f'최소값 : {min(numbers)} 최대값 : {max(numbers)}')

    # tuple() 은 하나의 인자만 받아야함 해서 매개변수들을 튜플로 선언해주고 사용함.

min_max(11, 19 ,1)
'''
 이외에도 방법 numbers(가변매개변수)를 가변매개변수로 여러 값을 받으면  튜플형태 받게됨.
def min_max(*numbers):
    print(type(numbers))
    print(f'최소값 : {min(numbers)} 최대값 : {max(numbers)}')

min_max(11, 19 ,1)
'''
```
