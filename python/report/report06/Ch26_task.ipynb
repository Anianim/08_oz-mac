0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

>제너레이터는 이너레이터를 생성한다. 이너레이터는 next() 함수로 호출가능하며 함수에 yield 값 이 있다면 yield 까지의 내용을 실행하고 yield 의 값을 return 한다.  

```python
def odd_numbers(n):
    for i in range(n+1): # n+1 까지해야 n을 포함한다.
        if i % 2 == 1:
            yield i #return은 한번만 리턴하지만 yield는 여러번 리턴 할 수 있다.

def add_number(): # 걸러낸 값을 출력하는 함수
    sum= []
    sum += odd_numbers(10)
    return sum

print(add_number())
```
