requests 모듈을 사용하여, 웹 페이지의 HTML을 가져오는 간단한 웹 스크레이퍼를 작성하세요.

'https://kream.co.kr/'에서<br>
데이터를 가져와서 상태 코드와 HTML의 첫 100자를 출력하세요.<br>
(실습 전 requests 모듈을 설치해야 함)

```python
import requests # requests 모듈을 가져옴

response = requests.get('https://kream.co.kr') #https://kream.co.kr/'에서 데이터를 가져옴. get() 은 가져오는 메서드
print("상태 코드:", response.status_code) # responese 는 가져온 값에 저장된 특정 값을 가져올때 사용함.
print("HTML 내용 (첫 100자):", response.text[:100]) #가져온 글자중 처음부터 100자 까지 출력 하기위한 구문
```
