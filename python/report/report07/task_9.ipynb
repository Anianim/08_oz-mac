````python
# animals/mammals.py

class Dog:
    def __init__(self, name):
        self.name = name

    def info(self):
        return f"Dog: {self.name}"

    def sound(self):
        return "멍멍"
````

````python
# nimals/birds.py

class Eagle:
    def __init__(self, name):
        self.name = name

    def info(self):
        return f"Eagle: {self.name}"

    def sound(self):
        return "샤악!"
````

````python
# main
from animals.mammals import Dog as my_dog #animals 패키지 안에 mammals 모듈 안의  Dog 클래스를 import함.
from animals.birds import Eagle as my_eagle

def main():
    print(my_dog.info())
    print("Sound:", my_dog.sound())

    print(my_eagle.info())
    print("Sound:", my_eagle.sound())

main()
````
