사용자 정의 예외 NegativeNumberError를 생성하고, 사용자로부터 숫자를 입력받아 음수일 경우 NegativeNumberError를 발생시키는 프로그램을 작성하세요.

- 예외 처리 구문을 사용하여 음수 입력 시 "양수를 입력해주세요."라는 메시지를 출력

```python
class NegativeNumberError(Exception): # 예외처리할 클래스 생성
    def __init__(self, notice) :
        self.except_name = "NegativeNumberError"
        self.notice = notice
    # __init__ 매직메서드 는 클래스 생성시에 실행 되는 함수이며 특징으로는 self를 첫 매개변수로 가진다.
    def __str__(self) : # 클래스 호출 될때 self를 매개변수로 받아 return
        return f'{self.except_name} : {self.notice}'

try:
    number = int(input("양수를 입력하세요: "))
    if number < 0:
        raise NegativeNumberError("음수가 입력되었습니다. 양수를 입력해주세요!") # 강제로 에러를 발생시키면서 문자열을 파라미터로 전달한다.
except NegativeNumberError as except_num: # 지정한 예외를 처리하기위한 구문
    print(except_num) # __str__ 메서드가 에러가 발생하면서 NegativeNumberError 의 매직메서드인 __str__ 으로 return 값을 받아온것을 출력함.
```
