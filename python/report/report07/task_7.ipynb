함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요.

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

```python
import time #time 모듈 가져와 사용

def timer_decorator(function):
    def wrapper():
        start_time = time.time() #시작 측정 strat
        result = function() #example_function() 이 실행됨
        end_time = time.time() # 끝시간 측정 end
        print(f"함수 실행 시간: {end_time - start_time}초") # end_time - start_time 뒤에 :f 포맷팅을 이용해서 소수점 자리를 정해서 출력가능하다 2f 소수점2자리까지 4f 4자리까지
        return result
    return wrapper

@timer_decorator
def example_function():
    sum = 0
    for i in range(1000000):
        sum += i
    return sum

example_function() # 호출이되면 데코레이터 되어있기때문에 timer_decorator() 함수가 먼저 호출되며 example_function()을 인자값으로 가지고 간다.
```
