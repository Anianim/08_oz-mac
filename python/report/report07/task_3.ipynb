사용자로부터 여러 개의 숫자를 입력받아 리스트에 저장한 후, 사용자가 'done'이라고 입력할 때까지 계속 입력받으세요.

- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



```python
def input_number(*numbers) : #초기 설정한 숫자들을 가변 매개변수로 받는 함수
    number_list = list(numbers) # 받은 파라미터를 리스트로 형변환 하고 담아준다.
    while True : # 숫자입력을 받기 위해 while문 사용
        num = (input("추가할 숫자를 입력 해주세요")) # 숫자를 입력받는다
        try: # 숫자라면 예외발생없이 내용실행
            if num != "done":
                number_list.append(int(num)) #문자라면 이부분에서 에러가 발생하게됨
            elif num == "done" :
                print(number_list)
                break
        except ValueError : # 예외를 처리 하는 부분
            print(f"잘못된 값: {num} 숫자를 입력해주세요!! ")

input_number(13 , 15 , 16 , 23)
```
