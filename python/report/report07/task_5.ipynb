datetime 모듈을 사용하여 현재 날짜와 시간을 출력하는 프로그램을 작성하세요.

또한, 사용자로부터 태어난 년도를 입력받아 현재 나이를 계산하여 출력하세요.

```python
from datetime import datetime  as date_time # datetime.datetime = as " "
now = date_time.now() #현재 날짜와 시간을 변수에 지정
print(f'현재 날짜와 시간 : {now}')

year = int(input("태어난 년도를 입력해주세요"))
age = now.year - year
print(f'현재 나이는 : {age}')
```
